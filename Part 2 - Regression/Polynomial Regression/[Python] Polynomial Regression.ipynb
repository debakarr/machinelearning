### Data Preprocessing

```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
%matplotlib inline
plt.rcParams['figure.figsize'] = [14, 8]

# Importing the dataset
dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:, 1:2].values
y = dataset.iloc[:, 2].values
```

We can see the non-linear relationship between the Salary and Level

### Fitting Linear Regression to the dataset

```python
lin_reg = LinearRegression()
lin_reg.fit(X, y)
```

### Fitting Polynomial Regression to the dataset

```python
poly_reg = PolynomialFeatures(degree = 2)
X_poly = poly_reg.fit_transform(X) # New matrix of feature
```

```python
X_poly
```

```python
# Include fit with poly_reg
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y)
```

### Visualising the Linear Regression results

```python
plt.scatter(X, y, c = 'red')
plt.plot(X, lin_reg.predict(X), c = 'green')
plt.title('Truth or Bluff (Linear Regression)')
plt.xlabel('Level')
plt.ylabel('Salary')
```

### Visualising the Polynomial Regression results

```python
plt.scatter(X, y, c = 'red')
plt.plot(X, lin_reg_2.predict(poly_reg.fit_transform(X)), c = 'green')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Level')
plt.ylabel('Salary')
```

### Polynomial Regression model with degree 3

```python
poly_reg = PolynomialFeatures(degree = 3)
X_poly = poly_reg.fit_transform(X) # New matrix of feature
```

```python
X_poly
```

```python
# Include fit with poly_reg
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y)
```

```python
plt.scatter(X, y, c = 'red')
plt.plot(X, lin_reg_2.predict(poly_reg.fit_transform(X)), c = 'green')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Level')
plt.ylabel('Salary')
```

### Polynomial Regression model with degree 4

```python
poly_reg = PolynomialFeatures(degree = 4)
X_poly = poly_reg.fit_transform(X) # New matrix of feature
```

```python
X_poly
```

```python
# Include fit with poly_reg
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y)
```

```python
plt.scatter(X, y, c = 'red')
plt.plot(X, lin_reg_2.predict(poly_reg.fit_transform(X)), c = 'green')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Level')
plt.ylabel('Salary')
```

### Get a more continuous curve

```python
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape((len(X_grid)), 1)
```

```python
X_grid[0:10, :] #Fist 10 elements
```

So X_grid contains 100 elements from 1 to 10 with step of 0.1

```python
plt.scatter(X, y, c = 'red')
plt.plot(X_grid, lin_reg_2.predict(poly_reg.fit_transform(X_grid)), c = 'green')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Level')
plt.ylabel('Salary')
```

### Predicting a new result with Linear Regression

```python
lin_reg.predict(6.5)
```

### Predicting a new result with Polynomial Regression

```python
lin_reg_2.predict(poly_reg.fit_transform(6.5))
```
